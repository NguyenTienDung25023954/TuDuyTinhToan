### W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không.
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
def luy_thua_cua_2(n):
    if n>0 and ((n & (n - 1)) == 0): #Nếu n là lũy thừa của 2
# n và n-1 không có chung bất kỳ vị trí bit 1 nào
# # nên & nhau chắc chắn bằng 0.
        print(f"{n} là một lũy thừa của 2.")
    else:
        print(f"{n} không phải là một lũy thừa của 2.")
n=int(input())
result = luy_thua_cua_2(n)
```

```python
def is_power_of_two(n):
    a=n
    if n<=0:
        print(f"{a} không phải là một lũy thừa của 2.")
    else:
        while n%2==0:
            n//=2
        if n==1:
            print(f"{a} là một lũy thừa của 2.")
        else:
            print(f"{a} không phải là một lũy thừa của 2.")
n=int(input())
result =is_power_of_two(n)
```

### W3A13
[Tính tiền điện]
Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 

```python
def tinh_tien_dien(so_dien):
    if so_dien <= 50:
        tien = so_dien * 1500
    elif so_dien <= 100:
        tien = 50 * 1500 + (so_dien - 50) * 2000
    else:
        tien = 50 * 1500 + 50 * 2000 + (so_dien - 100) * 3000
    print(f"Số tiền điện phải trả là: {tien} đồng")
so_dien=int(input())
result = tinh_tien_dien(so_dien)
```

### W3A17
Cấp số nhân
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng
thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs
d

```python
def cap_so_nhan(a,b,c,d):
    if a==0:
        if b==0 and c==0 and d==0:
            return "Yes"
        else:
            return "No"
    if b % a != 0:
        return "No"
    
    q = b // a 
    # Kiểm tra các số tiếp theo bằng phép nhân (Tránh lỗi chia cho 0)
    if b * q == c and c * q == d:
        return "Yes"
    else:
        return "No"
a, b, c, d = map(float, input().split())
result = cap_so_nhan(a,b,c,d)
print(result)
```

### Làm tròn đến số nhất định sau dấu phẩy

```python
def lam_tron(m,n):          #lam tron ma khong dung ham den chu so 3 sau dau phay
    n = n*(10**m)           #nhân n với 10 mũ m để dịch chuyển dấu phẩy thập phân sang phải m vị trí
    n += 0.5                #để làm tròn lên khi phần thập phân >=0.5, còn nếu <0.5 thì sẽ làm tròn xuống
    phan_du = n%1           #Lấy phần thập phân sau khi đã dịch chuyển
    phan_nguyen = n-phan_du #Lấy phần nguyên bằng cách trừ phần thập phân ra khỏi n đã dịch chuyển
    so_da_lam_tron = phan_nguyen / (10**m)  #Dịch chuyển dấu phẩy thập phân trở lại vị trí ban đầu bằng cách chia cho 10 mũ m
    print(so_da_lam_tron)
m = int(input())
n = float(input())
result = lam_tron(m,n)
```
