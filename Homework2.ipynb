### Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó. 

```python
def tinh_dien_tich_hinh_chu_nhat(chieu_dai, chieu_rong):
    dien_tich = chieu_dai * chieu_rong
    return dien_tich
chieu_dai, chieu_rong = map(float, input().split())
dien_tich = tinh_dien_tich_hinh_chu_nhat(chieu_dai, chieu_rong)
print(f"Dien tich hinh chu nhat: {dien_tich:.2f}")
```

### Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.

```python
import math
def tinh_dien_tich_hinh_tron(ban_kinh):
    dien_tich = math.pi * ban_kinh ** 2
    return dien_tich
ban_kinh = float(input())
dien_tich = tinh_dien_tich_hinh_tron(ban_kinh)  
print(f"Dien tich hinh tron: {dien_tich:.2f}")
```

### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.


```python
def kiem_tra_tam_giac(a,b,c):
    if a+b>c and b+c>a and a+c>b:
        p=(a+b+c)/2
        s=(p*(p-a)*(p-b)*(p-c))**0.5
        print(f"Dien tich tam giac: {s:.2f}")
        cv=a+b+c
        print(f"Chu vi tam giac: {cv:.2f}")
        if a==b==c:
            return "Tam giac deu"
        elif a==b or b==c or a==c:
            return "Tam giac can"
        elif a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2:
            return "Tam giac vuong"
        else:
            return "Tam giac thuong"
    else:
        return "Khong phai tam giac"
a, b, c = map(float, input().split())
ket_qua = kiem_tra_tam_giac(a, b, c)
print(ket_qua)
```

### Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
def giai_phuong_trinh_bac_2(a,b,c):
    if a==0:
        return f"Day khong la phuong trinh bac 2"
    else:
        delta=b**2 - 4*a*c
        if delta < 0:
            return f"Vo nghiem"
        elif delta == 0:
            x=-b/(2*a)
            return f"Phuong trinh co nghiem kep {x}"
        else:
            x1=(-b-(delta)**0.5)/(2*a)
            x2=(-b+(delta)**0.5)/(2*a)
            return f"Phuong trinh co 2 nghiem phan biet {x1} va {x2}"
a,b,c = map(float, input().split())
result=giai_phuong_trinh_bac_2(a,b,c)
print(result)
        
        
```

### Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím). 

```python
def tim_so_lon_nhat(m):
    max_num=0
    for i in m:
        if i > max_num:
            max_num=i
    return f"So lon nhat la {max_num}"
m=list(map(float,input().split()))
so_lon_nhat=tim_so_lon_nhat(m)
print(so_lon_nhat)
```

### Bài 8: Giải hệ phương trình tuyến tính:
##### ax + by = m
##### cx + dy = n

```python
def giai_he_phuong_trinh_bac_1(a,b,m,c,d,n):
    # vo nghiem hoac vo so nghiem 
    if a/c==b/d==m/n:
        return f"Phuong trinh vo so nghiem"
    elif a/c == b/d and (b/d != m/n or a/c != m/n):
        return f"Phuong trinh vo nghiem"
    else:
    # nhan c/a vao phuong trinh 1 roi tru 1 cho 2
        i = c/a
        y = (m*i - n)/(b*i - d)
        x = (m-b*y)/a
        return f"Phuong trinh co nghiem duy nhat x={x} va y={y}"
a,b,m,c,d,n=map(float,input().split())
result=giai_he_phuong_trinh_bac_1(a,b,m,c,d,n)
print(result)
```

### Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây

### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
def kiem_tra_vi_tri(a,b,r,x,y):
    khoang_cach=((x-a)**2+(y-b)**2)**0.5
    if khoang_cach==r:
        return f"A nam tren duong tron"
    elif khoang_cach>r:
        return f"A nam ngoai duong tron"
    else:
        return f"A nawm trong duong tron"
a,b=map(float,input("Nhap toa do tam: ").split())
r=float(input("Nhap ban kinh: "))
x,y=map(float,input("Nhap toa do A: ").split())
result=kiem_tra_vi_tri(a,b,r,x,y)
print(result)
```
