Bài 1: Nhập vào số n, in ra màn hình số gấp đôi n

```python
def bai1():
    n = int(input())
    m = 2 * n
    return m
bai1()
```

Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
def bai2():
    a, b = map(int, input().split())
    if b >= a:
        c = a/2
        s = a*b - 3.14*(c**2)
    else:
        c = b/2
        s = a*b - 3.14*(c**2)
    ket_qua = f"{s:.2f}"
    return ket_qua
bai2()
```

Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó.

```python
def bai3():
    c = input()
    if c.isupper():
        return c.lower()
    else:
        return c.upper()
bai3()
```

Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tự alphabet”, nếu sai in ra "{c}
không phải là kí tự alphabet".

```python
def bai4():
    c = input()
    if c.isalpha():
        return f"{c} là kí tự alphabet"
    else:
        return f"{c} không phải là kí tự alphabet"
bai4()
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 

```python
def bai5():
    chuhoa = input()
    stt=ord(chuhoa)-ord("A")
    vitri=(stt-1)%26
    chucantim=chr(vitri+ord("A"))
    ket_qua = chucantim.lower()
    return ket_qua
bai5()
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”

```python
def bai6():
    a, b, c = map(int,input().split())
    if a + b < c or a +c < b or b + c < a:
        ket_qua = "Không phải 3 cạnh của tam giác"
    else:
        p=(a+b+c)/2
        s=(p*(p-a)*(p-b)*(p-c))**0.5
        ket_qua = f"{s:.2f}"
    return ket_qua
bai6()
```

tương ứng (0 -> "Không", 1 -> "Một", ...)
