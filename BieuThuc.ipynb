```python
for i in range(5):
    row=""
    for j in range(5):
        if i==0 or j==2:
            row+="#"
        else:
            row+=" "
    row+=" "
    for j in range(5):
        if j==0 or (j<4 and (i==0 or i==4)) or (j==4 and 0<i<4):
            row+="#"
        else:   
            row+=" " 
    row+=" "
    for j in range(5):
        if i==0 or j==2:
            row+="#"
        else:
            row+=" "
    row+=" "
    for j in range(5):
        if i==0 or j==2:
            row+="#"
        else:
            row+=" "
    row+=" "
    print(row)
```
